```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df=pd.read_csv('epa-sea-level.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color='blue',marker='|')

plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level')
```

```python
slope, intercept ,_, _, _= linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Extend the line of best fit to the year 2050
future_years = range(1880, 2051)
plt.plot(future_years, slope * future_years + intercept, linestyle='dashed', label='first line of best fit')
```

```python
recent_data = df[df['Year'] >= 2000]
slope_recent, intercept_recent, _, _, _ = linregress(recent_data['Year'], recent_data['CSIRO Adjusted Sea Level'])

# Extend the new line of best fit to the year 2050
future_years = range(2000, 2051)
plt.plot(future_years, slope_recent * future_years + intercept_recent, linestyle='dashed', color='red', label='second line of best fit')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
```
